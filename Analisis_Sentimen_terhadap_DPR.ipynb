```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk

# Download VADER lexicon if not already
nltk.download('vader_lexicon')

# Initialize analyzer
sia = SentimentIntensityAnalyzer()

# Function to categorize sentiment
def get_sentiment(text):
    score = sia.polarity_scores(str(text))['compound']
    if score >= 0.05:
        return "Positive"
    elif score <= -0.05:
        return "Negative"
    else:
        return "Neutral"

# Apply to all reviews
df['sentiment'] = df['text'].apply(get_sentiment)

# Count frequency of each category
sentiment_counts = df['sentiment'].value_counts()
print(sentiment_counts)
```

```python
def get_sentiment_strict(text):
    score = sia.polarity_scores(str(text))['compound']
    if score > 0.1:
        return "Positive"
    else:
        return "Negative"

df['sentiment_strict'] = df['text'].apply(get_sentiment_strict)
print(df['sentiment_strict'].value_counts())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Extract reviews
reviews = df['text'].astype(str)

# Initialize TF-IDF with unigrams + bigrams (1-gram and 2-gram)
vectorizer = TfidfVectorizer(ngram_range=(1,2), stop_words='english')

# Fit and transform reviews into TF-IDF matrix
tfidf_matrix = vectorizer.fit_transform(reviews)

# Shape of the TF-IDF matrix (rows = reviews, columns = features)
print("TF-IDF matrix shape:", tfidf_matrix.shape)

# Example: feature names
print(vectorizer.get_feature_names_out()[:20])  # first 20 n-grams
```

```python
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Sum TF-IDF scores across all documents for each feature (keyword/ngram)
tfidf_scores = np.asarray(tfidf_matrix.sum(axis=0)).ravel()
feature_names = vectorizer.get_feature_names_out()

# Create mapping of feature -> score
keywords_scores = dict(zip(feature_names, tfidf_scores))

# Sort by score
sorted_keywords = sorted(keywords_scores.items(), key=lambda x: x[1], reverse=True)

# Get top 10 and bottom 10
top_10 = dict(sorted_keywords[:10])
bottom_10 = dict(sorted_keywords[-10:])

# Generate wordclouds
fig, axes = plt.subplots(1, 2, figsize=(14, 7))

# Top 10
wc_top = WordCloud(width=600, height=400, background_color='white').generate_from_frequencies(top_10)
axes[0].imshow(wc_top, interpolation='bilinear')
axes[0].set_title("Top 10 Keywords", fontsize=16)
axes[0].axis('off')

# Bottom 10
wc_bottom = WordCloud(width=600, height=400, background_color='white').generate_from_frequencies(bottom_10)
axes[1].imshow(wc_bottom, interpolation='bilinear')
axes[1].set_title("Lowest 10 Keywords", fontsize=16)
axes[1].axis('off')

plt.show()
```

```python
# 1. WordCloud dengan CountVectorizer (Unigram)
count_vect = CountVectorizer(stop_words='english')
X_count = count_vect.fit_transform(reviews)
word_freq_count = dict(zip(count_vect.get_feature_names_out(), X_count.sum(axis=0).A1))

wc_count = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq_count)

plt.figure(figsize=(10,5))
plt.imshow(wc_count, interpolation='bilinear')
plt.axis("off")
plt.title("WordCloud - CountVectorizer (Unigram)")
plt.show()
```
